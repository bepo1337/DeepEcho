```python
import pandas as pd
df = pd.read_json("../data/real_data_train.json")
```

```python
df_small = df.head(100)
df_small = df_small[["player_id", "validity_start", "league_minutes_played", "position", "club", "league"]]
df_small
```

```python
df_small.player_id.nunique()
```

```python
entity_columns = ["player_id"]
sequence_index = "validity_start"
context_columns = ["position"]
data_types = {
    "league_minutes_played" : "count",
    "club": 'categorical',
    "league": "categorical"
}
```

```python
from deepecho import PARModel

model = PARModel(epochs=1024, cuda=False)
model.fit(
    data=df_small,
    entity_columns=entity_columns,
    context_columns=context_columns,
    data_types=data_types,
    sequence_index=sequence_index,
)
model.sample(5)
```
